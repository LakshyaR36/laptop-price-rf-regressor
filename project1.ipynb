```python
def fetch(text):
    if text=='Intel Core i7' or text=='Intel Core i5' or text=='Intel Core i3':
        return text
    else:
        if text.split()[0]=='Intel':
            return 'Other Intel Processor'
        else:
            return 'AMD processor'
```

```python
df['Memory'] = df['Memory'].astype(str).replace(r'\.0' , '' , regex=True)
df['Memory'] = df['Memory'].str.replace('GB' , '')
df['Memory'] = df['Memory'].str.replace('TB' , '000')
new = df['Memory'].str.split("+" , n=1 , expand=True)
df["First"] = new[0]
df["First"] = df["First"].str.strip()
df["Second"] = new[1]
df['HDD1'] = df['First'].apply(lambda x: 1 if "HDD" in x else 0)
df['SSD1'] = df['First'].apply(lambda x: 1 if "SSD" in x else 0)
df['Hybrid1'] = df['First'].apply(lambda x: 1 if "Hybrid" in x else 0)
df['Flash1'] = df['First'].apply(lambda x: 1 if "Flash Storage" in x else 0)
df['First'] = df['First'].str.replace(r'\D' , '' , regex=True)
df['Second']=df['Second'].fillna("0")
df['HDD2'] = df['Second'].apply(lambda x:1 if "HDD" in x else 0)
df['SSD2'] = df['Second'].apply(lambda x:1 if "SSD" in x else 0)
df['Hybrid2'] = df['Second'].apply(lambda x:1 if "Hybrid" in x else 0)
df['Flash2'] = df['Second'].apply(lambda x: 1 if "Flash Storgae" in x else 0)
df['Second'] = df['Second'].str.replace(r'\D' , '' , regex=True)
df['First'] = df['First'].astype(int)
df['Second'] = df['Second'].astype(int)
df['HDD'] = (df['First']*df['HDD1']+df['Second']*df['HDD2'])
df['SSD'] = (df['First']*df['SSD1']+df['Second']*df['SSD2'])
df['Hybrid'] = (df['First']*df['Hybrid1']+df['Second']*df['Hybrid2'])
df['flash_storage'] = (df['First']*df['Flash1']+df['Second']*df['Flash2'])
```

```python
df.drop(columns=['First' , 'Second' , 'HDD1' , 'SSD1' , 'Hybrid1' , 'Flash1' , 'HDD2' , 'SSD2' , 'Hybrid2' , 'Flash2'] , inplace=True)
```

```python
x = df.drop(columns=['Price'])
y = np.log(df['Price'])
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , Y_train , Y_test = train_test_split(x,y,random_state=42,test_size=0.2) 
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
```

```python
from sklearn.linear_model import LinearRegression , Ridge , Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
```

```python
step1=ColumnTransformer(transformers=[
    ('trf1' , OneHotEncoder(sparse_output=False , drop='first') , [0,1,7,10,11])
] , remainder='passthrough')

step2 = LinearRegression()
pipe = Pipeline([
    ('step1' , step1) , 
    ('step2' , step2)
])

pipe.fit(X_train , Y_train)
Y_pred = pipe.predict(X_test)
```

```python
pipe
```

```python
from sklearn.metrics import r2_score , mean_absolute_error
print('R2 Score' , r2_score(Y_test , Y_pred))
print('MAE Score' , mean_absolute_error(Y_test , Y_pred))
```

```python
step1=ColumnTransformer(transformers=[
    ('trf1' , OneHotEncoder(sparse_output=False , drop='first') , [0,1,7,10,11])
] , remainder='passthrough')

step2 = RandomForestRegressor(
    n_estimators=300,
    random_state=3,
    max_samples=0.5,
    max_features=0.75,
    max_depth=15
)
pipe1 = Pipeline([
    ('step1' , step1) , 
    ('step2' , step2)
])

pipe1.fit(X_train , Y_train)
Y_pred = pipe1.predict(X_test)
```

```python
pipe1
```

```python
from sklearn.metrics import r2_score , mean_absolute_error
print('R2 Score' , r2_score(Y_test , Y_pred))
print('MAE Score' , mean_absolute_error(Y_test , Y_pred))
```
